```python
import pandas as pd
from collections import Counter, defaultdict
import glob
import json
import pandas as pd
import pigeonXT as pixt
```

```python
domains = pd.read_csv("../lite_data/domaintitles_annotated_no_international.csv", header=0)
```

```python
domains.head()
```

```python
annotations = pixt.annotate(
    domains, example_column='titleDEsc',
    options=['inrernational', 'domestic']
)
```

```python
domains.head()
```

```python
un_annoated  = domains.loc[domains['if_annotated_international']==False]
```

```python
annotations = pixt.annotate(
    un_annoated, example_column='titleDEsc',
    options=['inrernational', 'domestic']
)
```

```python
annotations.to_csv("../lite_data/international_intermediate_annotations.csv")
```

```python
annotations
```

```python
complete_annotations = annotations.loc[annotations['changed']==True]
new_annot_dict  = complete_annotations.set_index('domain').to_dict()['label']
new_annot_dict
```

```python
news_domains = domains.loc[domains['is_news']==1]
news_domains = news_domains.fillna('0')
news_domains['new_label'] = news_domains.apply(lambda row: row['international'] if row['international']!='0' else new_annot_dict[row['domain']], axis=1)
```
